```python
import numpy as np

def count_levels_either(category):
    # Missing/empty -> 0
    if pd.isna(category) or str(category).strip() == "":
        return 0
    s = str(category).strip()
    # choose separator (priority: '/' then ',')
    if '/' in s:
        parts = [p.strip() for p in s.split('/') if p.strip()]
    elif ',' in s:
        parts = [p.strip() for p in s.split(',') if p.strip()]
    else:
        parts = [s]  # single-level (no separator)
    return len(parts)

# Apply to both columns and get counts
category_hierarchy_counts_original = unique_epds['Category (original)'].apply(count_levels_either).value_counts().sort_index()
category_hierarchy_counts_english  = unique_epds['Category (en)'].apply(count_levels_either).value_counts().sort_index()

print("Product Category Hierarchy Counts (original):")
print(category_hierarchy_counts_original)
print("\nProduct Category Hierarchy Counts (en):")
print(category_hierarchy_counts_english)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1) Level counting function (either '/' or ',')
def count_levels_either(category):
    if pd.isna(category) or str(category).strip() == "":
        return 0
    s = str(category).strip()
    if '/' in s:
        parts = [p.strip() for p in s.split('/') if p.strip()]
    elif ',' in s:
        parts = [p.strip() for p in s.split(',') if p.strip()]
    else:
        parts = [s]
    return len(parts)

# 2) Add the level column to the DF
unique_epds = unique_epds.copy()                      # avoid modifying original accidentally
unique_epds['cat_levels'] = unique_epds['Category (original)'].apply(count_levels_either)

# If you want to count the english column instead, change the previous line to:
# unique_epds['cat_levels'] = unique_epds['Category (en)'].apply(count_levels_either)

# 3) Clean / normalize the Program operator column (optional)
unique_epds['Registration authority'] = unique_epds['Registration authority'].fillna('NAN').astype(str).str.strip()

# 4) Absolute counts: groupby operator x level
counts = unique_epds.groupby(['Registration authority', 'cat_levels']).size().unstack(fill_value=0).sort_index()
print("Absolute counts (rows = program operator, cols = number of hierarchy levels):")
print(counts)
# save to excel

counts.to_excel('product_category_hierarchy_counts_by_operator.xlsx')


```

#### Temporal Representativeness


```python
# EPDs temporal representativeness
unique_epds['Reference year'] = pd.to_numeric(unique_epds['Reference year'], errors='coerce')
unique_epds['Valid until'] = pd.to_numeric(unique_epds['Valid until'], errors='coerce')
valid_epds = unique_epds.dropna(subset=['Reference year', 'Valid until'])


valid_epds['Validity Duration'] = valid_epds['Valid until'] - valid_epds['Reference year']


total_valid_dates = len(valid_epds)

# 5-year validity
five_year = (valid_epds['Validity Duration'] == 5).sum()
five_year_pct = (five_year / total_valid_dates) * 100

# EPDs expired by end of 2024
expired_by_2025 = (valid_epds['Valid until'] <= 2025).sum()

# Specific counts for 3-year, 4-year, 6–7-year, 10-year, 15-year, 25-year...
duration_counts = valid_epds['Validity Duration'].value_counts()

# Anomalies: zero or negative validity duration
anomalies = valid_epds[valid_epds['Validity Duration'] <= 0]

# Print results
print(f"Total EPDs with valid date info: {total_valid_dates}")
print(f"EPDs with 5-year validity: {five_year} ({five_year_pct:.1f}%)")
print(f"EPDs expired by end of 2025: {expired_by_2025}")
print("\nValidity duration counts (years):")
print(duration_counts.sort_index())

print(f"\nEPDs with zero or negative validity duration: {len(anomalies)}")

```
