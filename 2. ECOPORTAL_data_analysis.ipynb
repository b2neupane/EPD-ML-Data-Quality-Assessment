```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# --- Helper functions ---
def _clean_category(series: pd.Series) -> pd.Series:
    """Normalize a category series: strip whitespace, collapse blanks, and map common 'unknown' tokens to empty."""
    s = series.copy()
    s = s.astype(str)
    s = s.replace({np.nan: ''})
    s = s.str.strip()
    s = s.str.replace(r'\s+', ' ', regex=True)
    s = s.str.replace(r'^[-–—\.]+$', '', regex=True)  # lone punctuation -> blank
    
    # common tokens signalling missing info (multi-lingual where possible)
    missing_tokens = {
        '', 'nan', 'none', 'null', 'n/a', 'na', 'n.a.', 'not available', 'not applicable',
        'unknown', 'unspecified', 'unclassified', 'keine angabe', 'ohne angabe',
        'sconosciuto', 'indéfini', 'indefinido', 'sin categoría', 'aucune'
    }
    s_lower = s.str.lower()
    s = s.mask(s_lower.isin(missing_tokens), '')
    return s

def _safe_col(df: pd.DataFrame, name: str) -> pd.Series:
    """Return a column by name if present; else return an empty series of blanks."""
    if name in df.columns:
        return df[name]
    else:
        return pd.Series([''] * len(df))

def _plot_top_counts(counts: pd.Series, title: str, outfile: str, xlabel: str = "Count", topn: int = 15):
    """Create a horizontal bar chart of top-N counts."""
    if counts.empty:
        print(f"No data to plot for: {title}")
        return None
    c = counts.sort_values(ascending=True).tail(topn)  # small-to-large for horizontal bar
    plt.figure(figsize=(9, 5.5))
    y = np.arange(len(c))
    plt.barh(y, c.values)
    plt.yticks(y, c.index)
    plt.xlabel(xlabel)
    plt.title(title)
    plt.grid(axis='x', linestyle='--', alpha=0.5)
    # annotate values
    for i, v in enumerate(c.values):
        plt.text(v + max(c.values) * 0.01, i, str(v), va='center')
    plt.tight_layout()
    plt.savefig(outfile, dpi=300, bbox_inches='tight')
    plt.show()
    return outfile

# --- Main analysis ---
# Expecting a DataFrame named `unique_epds` with columns:
# 'Category (en)', 'Category (original)', 'Registration authority'
if 'unique_epds' not in globals():
    print("DataFrame `unique_epds` not found in the environment. "
          "Please ensure it is loaded and contains the columns: "
          "'Category (en)', 'Category (original)', and 'Registration authority'.")
else:
    df = unique_epds.copy()

    # Clean and normalize category fields
    cat_en = _clean_category(_safe_col(df, 'Category (en)'))
    cat_orig = _clean_category(_safe_col(df, 'Category (original)'))
    df['_cat_en_clean'] = cat_en
    df['_cat_orig_clean'] = cat_orig

    # Missing masks
    miss_en = df['_cat_en_clean'] == ''
    miss_orig = df['_cat_orig_clean'] == ''
    miss_both = miss_en & miss_orig
    miss_either = miss_en | miss_orig

    total = len(df)
    summary = pd.DataFrame({
        'Count': [
            miss_en.sum(),
            miss_orig.sum(),
            miss_both.sum(),
            (~miss_either).sum()
        ],
        'Percent': [
            miss_en.mean() * 100,
            miss_orig.mean() * 100,
            miss_both.mean() * 100,
            ((~miss_either).mean() * 100)
        ]
    }, index=['Missing EN only or EN missing', 'Missing Original only or Original missing',
              'Missing Both (EN & Original)', 'Present in Both']).round(2)

    # Display summary table (try pretty display if available)
    try:
        from ace_tools import display_dataframe_to_user
        display_dataframe_to_user("Category Missingness Summary", summary.reset_index().rename(columns={'index':'Status'}))
    except Exception as e:
        print("Summary table:\n", summary)

    # Missing by programme operator
    prog_col = 'Registration authority' if 'Registration authority' in df.columns else None
    if prog_col:
        topN = 15
        en_by_prog = df.loc[miss_en, prog_col].value_counts()
        orig_by_prog = df.loc[miss_orig, prog_col].value_counts()
        both_by_prog = df.loc[miss_both, prog_col].value_counts()

        # Save plots
        out1 = _plot_top_counts(en_by_prog, f"Top {topN} programme operators with missing Category (EN)", 
                                "/mnt/data/missing_category_en_by_program.png", xlabel="EPDs with missing EN category", topn=topN)
        out2 = _plot_top_counts(orig_by_prog, f"Top {topN} programme operators with missing Category (Original)", 
                                "/mnt/data/missing_category_orig_by_program.png", xlabel="EPDs with missing original category", topn=topN)
        out3 = _plot_top_counts(both_by_prog, f"Top {topN} programme operators with missing Category in BOTH fields", 
                                "/mnt/data/missing_category_both_by_program.png", xlabel="EPDs missing both categories", topn=topN)

        # Also provide a combined table of counts by operator
        combined = (pd.DataFrame({
            'Missing EN': en_by_prog,
            'Missing Original': orig_by_prog,
            'Missing Both': both_by_prog
        }).fillna(0).astype(int)
           .sort_values('Missing Both', ascending=False))

        try:
            from ace_tools import display_dataframe_to_user
            display_dataframe_to_user("Missing Categories by Programme Operator", combined.reset_index().rename(columns={'index': prog_col}))
        except Exception:
            print("\nMissing categories by programme operator (top rows):\n", combined.head(20))

        print("\nFiles saved:")
        print(out1, out2, out3)
```

```python
# Reference Unit Analysis
ref_units = unique_epds['Ref. unit'].fillna('').str.strip()
total_epds = len(ref_units)
unique_units = ref_units.unique()
num_unique_units = len(unique_units)
print(f"Unique unit strings: {num_unique_units}\n")

standard_units = {
    'mass': ['kg', 'ton', 'lb'],
    'volume': ['m3', 'm³'],
    'area': ['m2', 'm²'],
    'length': ['m', 'mm', 'meter'],
    'count': ['unit', 'piece', 'pieces']
}

all_standard_units = [u.lower() for sublist in standard_units.values() for u in sublist]

unit_analysis = ref_units.str.lower().value_counts().reset_index()
unit_analysis.columns = ['unit', 'count']

def classify_unit(unit):
    for cat, valid_units in standard_units.items():
        if unit in [u.lower() for u in valid_units]:
            return cat
    return 'nonstandard'

unit_analysis['category'] = unit_analysis['unit'].apply(classify_unit)

standard_count = unit_analysis[unit_analysis['category'] != 'nonstandard']['count'].sum()
nonstandard_pct = (unit_analysis[unit_analysis['category'] == 'nonstandard']['count'].sum() / total_epds) * 100
nonstandard_count = unit_analysis[unit_analysis['category'] == 'nonstandard']['count'].sum()

print(f"Standard unit matches: {standard_count}")
print(f"Nonstandard unit entries: {nonstandard_count}")
print(f"Percentage of nonstandard units: {nonstandard_pct:.1f}%")
```

```python
import re
ref_units = unique_epds['Ref. unit'].fillna('').str.strip().str.lower()
unit_analysis = ref_units.value_counts().reset_index()
unit_analysis.columns = ['unit', 'count']

standard_patterns = [
    r'^kg$', r'^ton$', r'^lb$', r'^m3$', r'^m²$', r'^m2$', r'^m³$', r'^m$', r'^mm$', r'^meter$',
    r'^unit$', r'^piece$', r'^pieces$'
]

def is_standard(unit):
    return any(re.match(pattern, unit) for pattern in standard_patterns)

unit_analysis['is_suspicious'] = unit_analysis['unit'].apply(
    lambda u: not is_standard(u) and not re.match(r'^\d+\s*[a-zA-Z]+$', u) and not u.isnumeric()
)
suspicious_units = unit_analysis[unit_analysis['is_suspicious']]
print("\nSuspicious units:")
print(suspicious_units[['unit', 'count']])
```

```python
# Filter for A2 EPDs
a2_epds = unique_epds[unique_epds['Compliance'].str.contains('15804\+a2', case=False, na=False)]

# Find all A2 indicator columns
a2_indicator_cols = [col for col in unique_epds.columns if '(A2)' in col]

# Find all life cycle stage columns (e.g., A1-A3, A4, etc.)
lifecycle_stages = ['A1-A3', 'A4', 'A5', 'B1', 'B4', 'B5', 'C1', 'C2', 'C3', 'C4', 'D']

# Prepare a table: rows=indicators, columns=module_present/module_missing for each stage
columns = []
for stage in lifecycle_stages:
    columns.append(f'{stage}_Present')
    columns.append(f'{stage}_Missing')

indicator_stage_table = pd.DataFrame(index=a2_indicator_cols, columns=columns)

def count_present_missing(df, indicator, stage):
    mask = df['Module'].str.contains(stage, na=False)
    present = df[mask & df[indicator].notna() & (df[indicator] != '')]['UUID'].nunique()
    total = df[mask]['UUID'].nunique()
    missing = total - present
    return present, missing

for indicator in a2_indicator_cols:
    for stage in lifecycle_stages:
        present, missing = count_present_missing(a2_epds, indicator, stage)
        indicator_stage_table.loc[indicator, f'{stage}_Present'] = present
        indicator_stage_table.loc[indicator, f'{stage}_Missing'] = missing

indicator_stage_table = indicator_stage_table.fillna(0).astype(int)

print("Table: Unique A2 EPDs with environmental indicator data by life cycle stage (separate Present/Missing columns)")
display(indicator_stage_table)

# save indicator stage table to excel
indicator_stage_table.to_excel('indicator_stage_table.xlsx')
```
